```python
(trainingData, testData) = spark_df.randomSplit([0.7, 0.3])
```

```python
trainingData.count()
```

```python
testData.count()
```

```python
trainingData.cache()
testData.cache()
print("We have %d training instances and %d test instances." % (trainingData.count(), testData.count()))
```

```python
from pyspark.ml.regression import DecisionTreeRegressor, DecisionTreeRegressionModel

dt = DecisionTreeRegressor(labelCol="COGT", featuresCol="features")
```

```python
from pyspark.ml.feature import VectorAssembler

# Use Vector Assembler to create Feature Vector
# We wil pass these into our Pipeline later

assembler = VectorAssembler(inputCols=["T", "RH", "AH", "PT08S1CO", "NMHCGT", "C6H6GT", "PT08S2NMHC", "NOxGT", "PT08S3NOx", "NO2GT", "PT08S4NO2", "PT08S5O3"], outputCol="features")
```

```python
from pyspark.ml import Pipeline

dtPipeline = Pipeline(stages = [assembler, dt])

#Create initial decision tree model
initialModel = dtPipeline.fit(trainingData)
```

```python
initialPredictions = initialModel.transform(testData)
```

```python
selected = initialPredictions.select("COGT", "prediction")
```

```python
display(selected.toPandas())
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

# Evaluate model using the RegressionEvaluator
evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="COGT")
evaluator.evaluate(initialPredictions)
```

```python
evaluator.getMetricName()
```

```python
print evaluator.explainParams()
```

```python
# Split the trainingData into 2 parts: partTrainingData, validationData
(partTrainingData, validationData) = trainingData.randomSplit([0.7, 0.3])
partTrainingData.cache()
validationData.cache()
print 'Tuning model using %d training instances and %d validation instances' % (partTrainingData.count(), validationData.count())
```

```python
# We will try these settings for maxDepth
maxDepthSettings = [1, 2, 4, 6, 8, 10, 12, 14]

# Function that creates models for every depth in maxDepthSettings
def createModel(depth, data):
  dt = DecisionTreeRegressor(labelCol="COGT", featuresCol="features", maxDepth=depth)
  dtPipeline = Pipeline(stages = [assembler, dt])
  dtModel = dtPipeline.fit(data)
  return dtModel

# Create models
models = map(lambda depth: createModel(depth, partTrainingData), maxDepthSettings)

# Create predictions
trainPredictions = map(lambda model: model.transform(partTrainingData), models)
valPredictions = map(lambda model: model.transform(validationData), models)

# Calculate RMSE
trainRMSE = map(lambda prediction: evaluator.evaluate(prediction), trainPredictions)
valRMSE = map(lambda prediction: evaluator.evaluate(prediction), valPredictions)
```

```python
# Plot the training and validation errors for the trees of different depths.
from matplotlib import pyplot
pyplot.clf()
pyplot.plot(maxDepthSettings, trainRMSE, 'b-', label='training error')
pyplot.hold(True)
pyplot.plot(maxDepthSettings, valRMSE, 'r-', label='validation error')
pyplot.xlabel('maxDepth')
pyplot.ylabel('error (RMSE)')
pyplot.ylim(ymin=0)
pyplot.legend()
pyplot.title('Tuning maxDepth: Evaluation and Overfitting')
display()
```

```python
import numpy
bestMaxDepth = maxDepthSettings[numpy.argmin(valRMSE)]
print 'Chose best maxDepth = %d' % bestMaxDepth
```

```python
# Train a tree using ALL of the trainingData and using the BEST maxDepth parameter we found.
dt = DecisionTreeRegressor(labelCol="COGT", featuresCol="features", maxDepth=bestMaxDepth)
dtPipeline = Pipeline(stages = [assembler, dt])
finalModel = dtPipeline.fit(trainingData)
finalTrainPredictions = finalModel.transform(trainingData)
finalTestPredictions = finalModel.transform(testData)
```

```python
print finalModel.stages[1]
```

```python
finalTrainRMSE = evaluator.evaluate(finalTrainPredictions)
finalTestRMSE = evaluator.evaluate(finalTestPredictions)

print 'Final model achieves:'
print '  Training error: %g' % finalTrainRMSE
print '  Test error:     %g' % finalTestRMSE
```
