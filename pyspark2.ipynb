# Analysis of defaults of credit cards clients 

```python
## Download and import the file
import pandas as pd
import xlrd 
import urllib2

link ='https://github.com/adrian-dembek/CreditDefault/raw/master/default%20of%20credit%20card%20clients.xls'
socket = urllib2.urlopen(link)

xd = pd.ExcelFile(socket)
```

```python
df = xd.parse(xd.sheet_names[-1], header=1, index_col = 0)
```

## Create Spark Data Frame using sqlContext

```python
spark_df = sqlContext.createDataFrame(df)
```

```python
spark_df.printSchema()
```

```python
from IPython.core.display import display
display(spark_df.toPandas().head())
```

```python
spark_df = spark_df.withColumnRenamed("default payment next month", "default")
```

```python
## Change type of variable
#from pyspark.sql.types import DoubleType
#spark_df = spark_df.withColumn("default", spark_df.default.cast(DoubleType()))

from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol="default", outputCol="defaultIndex")
spark_df = indexer.fit(spark_df).transform(spark_df)

indexer = StringIndexer(inputCol="SEX", outputCol="sexIndex")
spark_df = indexer.fit(spark_df).transform(spark_df)

indexer = StringIndexer(inputCol="EDUCATION", outputCol="educationIndex")
spark_df = indexer.fit(spark_df).transform(spark_df)

indexer = StringIndexer(inputCol="MARRIAGE", outputCol="marriageIndex")
spark_df = indexer.fit(spark_df).transform(spark_df)

display(spark_df.toPandas())
```

```python
from pyspark.ml.feature import OneHotEncoder
```

```python
encoder = OneHotEncoder(inputCol="educationIndex", outputCol="educationOne")
spark_df = encoder.transform(spark_df)

encoder = OneHotEncoder(inputCol="marriageIndex", outputCol="marriageOne")
spark_df = encoder.transform(spark_df)
```

```python
## Remove columns

#spark_df = spark_df.drop("SEX", "EDUCATION", "MARRIAGE", "default", "educationIndex", "marriageIndex")

spark_df = spark_df.select([c for c in spark_df.columns if c not in {"SEX", "EDUCATION", "MARRIAGE", "default", "educationIndex", "marriageIndex"}])
```

## Format variables

```python
#stringIndexer = StringIndexer(inputCol="label", outputCol="indexed")
```

```python
from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(inputCols=spark_df.columns[1:23], outputCol="features")
```

```python
#from pyspark.ml.feature import VectorIndexer

# Automatically identify categorical features, and index them.
# Set maxCategories so features with > 4 distinct values are treated as continuous.
#featureIndexer =\
#    VectorIndexer(inputCol="features", outputCol="indexedFeatures", maxCategories=6)
    #.fit(spark_df)
```

```python
spark_df = assembler.transform(spark_df)
```

```python
display(spark_df.toPandas())
```

## Create training and testing data

```python
(trainingData, testData) = spark_df.randomSplit([0.7,0.3])
trainingData.cache()
testData.cache()
```

## Decision tree basic model

```python
from pyspark.ml.regression import DecisionTreeRegressor
from pyspark.ml.classification import DecisionTreeClassifier

dtc = DecisionTreeClassifier(labelCol="defaultIndex", featuresCol="features")
```

```python
from pyspark.ml import Pipeline

dtPipeline = Pipeline(stages = [assembler, dtc])
```

```python
dectree = dtc.fit(trainingData)
```

```python
dectree.featureImportances
```

### Classification tree

```python
#Create initial decision tree model
initialModel = dtPipeline.fit(trainingData)
```

```python
initialPredictions = initialModel.transform(testData)
```

```python
selected = initialPredictions.select("defaultIndex", "prediction")
```

```python
initialPredictions.select("prediction")
```

```python
display(selected.toPandas())
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

# Evaluate model using the RegressionEvaluator
evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="defaultIndex")
evaluator.evaluate(initialPredictions)
```

### Evaluate

```python
treeModel = initialModel.stages[1]
```

```python
print(treeModel)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

predictions = initialModel.transform(testData)

# Select example rows to display.
predictions.select("prediction", "defaultIndex", "features").show(5)

# Select (prediction, true label) and compute test error
evaluator = MulticlassClassificationEvaluator(
    labelCol="defaultIndex", predictionCol="prediction", metricName="precision")
accuracy = evaluator.evaluate(predictions)
print("Test Error = %g" % (1.0 - accuracy))
```
